```python
from youtube_transcript_api import YouTubeTranscriptApi 
  
# assigning srt variable with the list 
# of dictionaries obtained by the get_transcript() function
yt_vid_id = "Ko5Lnkt2sTg"
transcript_list = YouTubeTranscriptApi.list_transcripts(yt_vid_id)
# print(transcript_list)
# srt = YouTubeTranscriptApi.get_transcript("Ko5Lnkt2sTg", languages=['ar'])
```

```python
for transcript in transcript_list:
    print(transcript.language_code)
    if transcript.language_code == "ar":
        subtitile = transcript.fetch()
        break

    else:
        subtitile = transcript.translate('ar').fetch()
        break
    
with open("ar_subtitles.srt", "w") as file:
    for i, sub in enumerate(subtitile):
        file.write(f"{i+1}\n")
        file.write(f"{sub['start']} --> {sub['start']+sub['duration']}\n")
        file.write(f"{sub['text']}\n\n")
with open("ar_transcript.txt", "w") as file:
    for i, sub in enumerate(subtitile):
        file.write(f"{sub['text']}\n")
```
